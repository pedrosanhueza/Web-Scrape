# Name and Period

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.senado.cl/appsenado/index.php?mo=senadores&ac=periodos'

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
senadores = soup.find_all('tr')[1:]

rows=[]

for senador in senadores:
    row={}
    row['Nombre'] = senador.find('td').text.strip()
    row['Periodos'] = senador.find_all('td')[-1].text[2:]
    rows.append(row)
```

```python
data = pd.DataFrame(rows)
```

# Session and Theme

```python
rows=[]
for votaid in range(8920,8951):
    url = f'https://www.senado.cl/appsenado/index.php?mo=sesionessala&ac=detalleVotacion&votaid={votaid}' # max 8950 min 286
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    row={}
    encabezado = soup.find('div',{'class':'col1'}).text.split('\n')[1]
    row['Sesion'] = encabezado.split(' Tema: ')[0][8:]
    try:
        row['Tema'] = encabezado.split(' Tema: ')[1]
    except:
        pass
    rows.append(row)
```

```python
data = pd.DataFrame(rows)
```

# Name and Votes

```python
votaid = 8951
url = f'https://www.senado.cl/appsenado/index.php?mo=sesionessala&ac=detalleVotacion&votaid={votaid}' # max 8950 min 286 or min 8633
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
senadores = soup.find_all('tr')[1:]
```

```python
rows=[]
for senador in senadores:
    row={}
    row['Senador'] = senador.find('td').text
    if 'gif' in str(senador.find_all('td')[1]):
        row['Si'] = 'Si'
    else:
        row['Si'] = ''
    if 'gif' in str(senador.find_all('td')[2]):
        row['No'] = 'No'
    else:
        row['No'] = ''
    if 'gif' in str(senador.find_all('td')[3]):
        row['Abstencion'] = 'Abstencion'
    else:
        row['Abstencion'] = ''
    if 'gif' in str(senador.find_all('td')[4]):
        row['Pareo'] = 'Pareo'
    else:
        row['Pareo'] = ''
    rows.append(row)
```

```python
data = pd.DataFrame(rows)
```

```python
data
```

# Name and Party

```python
url = f'https://www.senado.cl/appsenado/index.php?mo=senadores&ac=listado'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
senadores = soup.find_all('tr')[1::4]

rows = []

for senador in senadores:
    row = {}
    row['Name'] = senador.find_all('td')[2].text.strip().split('\n')[0]
    row['Partido'] = senador.find_all('td')[3].text[9:]
    row['Region'] = senador.find_all('td')[2].text.strip().split('\n')[1].strip().split(' | ')[0][8:]
    row['Telefono'] = senador.text.strip()[senador.text.strip().find('Teléfono')+10:senador.text.strip().find('Teléfono')+25]
    row['Circunscripcion'] = senador.text.strip()[senador.text.strip().find('Circunscripción:')+17:senador.text.strip().find('Circunscripción:')+19].strip()
    row['Image'] = 'https://www.senado.cl'+senador.find('img')['src']
    rows.append(row)
```

```python
data = pd.DataFrame(rows)

data
```
