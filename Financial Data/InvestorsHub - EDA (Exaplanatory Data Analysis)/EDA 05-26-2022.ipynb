```python
# create a list with websites containing financials data
letters = [ chr(x) for x in range(65,91)] # list with A - Z letters
url_NASDAQ = [ "https://www.advfn.com/nasdaq/nasdaq.asp?companies=" + letter for letter in letters ] # NASDAQ
url_NYSE   = [ "https://www.advfn.com/nyse/newyorkstockexchange.asp?companies=" + letter for letter in letters ] # NYSE
url_AMEX   = [ "https://www.advfn.com/amex/americanstockexchange.asp?companies=" + letter for letter in letters ] # AMEX

len(url_NASDAQ + url_NYSE + url_AMEX)
```

```python
# download html from url
def getSoup(url):
    request=urllib.request.Request(url,None,headers)
    response = urllib.request.urlopen(request)
    data = response.read()
    soup = BeautifulSoup(data, 'html.parser')
    return soup
```

```python
# store half of the financial links - part 1
def getLinks1(soup_links, list_name):
    for e in soup_links.find_all('tr', {"class":"ts0"}):
        for links in e:
            for link in links.find_all("a"):
                if link.get("href").endswith("financials"):
                    list_name.append(link.get("href"))
    return list_name
```

```python
# store half of the financial links - part 2
def getLinks2(soup_links, list_name):
    for e in soup_links.find_all('tr', {"class":"ts1"}):
        for links in e:
            for link in links.find_all("a"):
                if link.get("href").endswith("financials"):
                    list_name.append(link.get("href"))
    return list_name
```

```python
# Takes about 140 seconds
# RUN ALL FUNCTIONS

# Collect all NASDAQ financial statements links
link_NASDAQ = []
for letter in url_NASDAQ:
    soupLetter = getSoup(letter)
    getLinks1(soupLetter, link_NASDAQ)
    getLinks2(soupLetter, link_NASDAQ)
print(len(link_NASDAQ), " from NASDAQ") # TRACKING amount

# Collect all NYSE financial statements links
link_NYSE = []
for letter in url_NYSE:
    soupLetter = getSoup(letter)
    getLinks1(soupLetter, link_NYSE)
    getLinks2(soupLetter, link_NYSE)
print(len(link_NYSE), " from NYSE") # TRACKING amount

# Collect all AMEX financial statements links
link_AMEX = []
for letter in url_AMEX:
    soupLetter = getSoup(letter)
    getLinks1(soupLetter, link_AMEX)
    getLinks2(soupLetter, link_AMEX)
print(len(link_AMEX), " from AMEX") # TRACKING amount
```
