```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://liquidations.walmart.com/?limit=128'

response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text)
```

```python
items = soup.find_all('div',{'class','product-info'})
```

```python
rows = []
for item in items:
    row={}

    row['data-auction-id'] = item.find('span',{'class','countdown'})['data-auction-id']
    row['product-name'] = item.find('div',{'class','product-name'}).text.strip()
    row['current_bid'] = item.find('div',{'class','current_bid'}).text.strip()[14:].replace(',','')
    row['cost_per_unit'] = item.find('div',{'class','cost_per_unit'}).text.strip()[64:].replace(',','')
    row['bids_number'] = item.find('div',{'class','bids_number'}).text.strip()[6:]
    row['data-end-time'] = item.find('span',{'class','countdown'})['data-end-time']
    row['product-info'] = item.find('div',{'class','product-name'}).find('a')['href']

    rows.append(row)
```

```python
data = pd.DataFrame(rows)
```

```python
data.to_csv('walmart_liquidations.csv', index=False)
```
