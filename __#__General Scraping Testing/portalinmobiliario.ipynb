```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
operaciones = [
    'venta',
    'arriendo',
    'arriendo-de-temporada'
]

propiedades = [
    'departamento',
    'casa',
    'oficina',
    'parcela',
    'comercial',
    'terreno-en-construccion',
    'sitio',
    'bodega',
    'agricola',
    'otros',
    'estacionamiento',
    'loteo'
]
```

```python
page_index = ['_Desde_1'] + ['_Desde_' + str(x * 50 + 1) for x in range(1,41)]

urls = [] # list of all links

for operacion in operaciones:

    for propiedad in propiedades:

        page = ''

        for idx, page in enumerate(page_index):

            url = f'https://www.portalinmobiliario.com/{operacion}/{propiedad}/{page}_OrderId_BEGINS*DESC_NoIndex_True'

            page = f'{page_index[idx]}'

            urls.append(url)
```

```python
rows = []

def append_rows(url):

    response = requests.get(url)

    soup = BeautifulSoup(response.text)

    rents = soup.find_all('li',{'class','ui-search-layout__item shops__layout-item'})

    for rent in rents:
        
        row={}
        
        row['Precio'] = rent.find('span',{'class','price-tag-text-sr-only'}).text
        
        try:
            row['Metros cuadrados'] = rent.find('li',{'class','ui-search-card-attributes__attribute'}).text
        except:
            pass
        
        try:
            row['Dormitorios'] = rent.find_all('li',{'class','ui-search-card-attributes__attribute'})[1].text
        except:
            pass
        
        try:
            row['Banos'] = rent.find_all('li',{'class','ui-search-card-attributes__attribute'})[2].text
        except:
            pass
        
        row['Tipo de propiedad'] = rent.find('span',{'class','ui-search-item__group__element ui-search-item__subtitle shops__items-group-details'}).text
        
        row['Ubicacion'] = rent.find('p',{'class','ui-search-item__group__element ui-search-item__location shops__items-group-details'}).text
        
        row['url'] = url
        
        rows.append(row)

    # break

for idx, url in enumerate(urls):
    try:
        append_rows(url)
        print(f'{idx}/{len(urls)} completed:',url)
    except:
        print(f'{idx}/{len(urls)} broke: ',url)

```

```python
data = pd.DataFrame(rows)
```

```python
data.to_csv('portalinmobiliario_venta.csv', index=False)
```
